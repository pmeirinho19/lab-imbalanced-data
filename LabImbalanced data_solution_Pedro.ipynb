```python
import pandas as pd
import numpy as np
```

```python
# 1. Load the dataset and explore the variables.

customer_churn_df = pd.read_csv('customer_churn.csv')
customer_churn_df
```

```python
#2. We will try to predict variable Churn using a logistic regression on variables tenure, SeniorCitizen,MonthlyCharges.
    #(just to inform?)

#3. Extract the target variable

customer_churn_df['Churn']
customer_churn_df['Churn'].value_counts() # --> Binary clasification problem

y = customer_churn_df['Churn']
```

```python
4# Extract the independent variables and scale them.

customer_churn_df['tenure'] #--> To use MinMax Scaler
#display(customer_churn_df['tenure'].value_counts())

customer_churn_df['SeniorCitizen']# --> categorical, not possible to be scaled, it will be treated as a numerical one and his values will be kept
#display(customer_churn_df['SeniorCitizen'].value_counts())

customer_churn_df['MonthlyCharges']  #--> To use MinMax Scaler
#display(customer_churn_df['MonthlyCharges'].value_counts())

X = customer_churn_df.loc[:,['tenure','SeniorCitizen', 'MonthlyCharges']]

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 0)


from sklearn.preprocessing import MinMaxScaler

transformer = MinMaxScaler()     #Doubt: Can we do the scalers before the train-test split?
transformer.fit(X_train)

X_train_scaled1 = transformer.transform(X_train)
X_train_scaled = pd.DataFrame(X_train_scaled1, columns = X_train.columns, index=X_train.index)

X_test_scaled1 = transformer.transform(X_test)
X_test_scaled = pd.DataFrame(X_test_scaled1, columns = X_test.columns, index=X_test.index)
```

```python
# 5. Build the logistic regression model.

from sklearn.linear_model import LogisticRegression 

classification = LogisticRegression(random_state=0, solver='lbfgs', multi_class = 'ovr' )
classification.fit(X_train_scaled, y_train)

y_train_pred = classification.predict(X_train_scaled)
y_test_pred = classification.predict(X_test_scaled)

```

```python
#6 Evaluate the model

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score 

print("The accuracy in the TRAIN set is: {:.3f}".format(accuracy_score(y_train, y_train_pred)))
print("The accuracy in the TEST  set is: {:.3f}".format(accuracy_score(y_test, y_test_pred)))
print('\n')
print("The precission in the TRAIN set is: {:.3f}".format(precision_score(y_train, y_train_pred, pos_label="Yes")))
print("The precission in the TEST  set is: {:.3f}".format(precision_score(y_test, y_test_pred, pos_label="Yes")))
print('\n')
print("The recall in the TRAIN set is: {:.3f}".format(recall_score(y_train, y_train_pred, pos_label="Yes")))
print("The recall in the TEST  set is: {:.3f}".format(recall_score(y_test,  y_test_pred, pos_label="Yes")))
print('\n')
print("The F1-score for the TRAIN set is {:.2f}".format(f1_score(y_train,y_train_pred, pos_label="Yes")))
print("The F1-score for the TEST set is {:.2f}".format(f1_score(y_test,y_test_pred, pos_label="Yes")))
```

```python
#7. Even a simple model will give us more than 70% accuracy. Why?

#R: Because we have much more samples of "Yes" in 'Churn' column. 
    #Therefore, the model can predict very well the true positives (TP) that is 'Churn' = Yes.
    # we have to keep in mind that the accuracy formula is =(TP+TN)/(TP+TN+FP+FN)
    # If we only see the accuracy we are not able to say if our model is good or bad.
```

```python
#8.  Synthetic Minority Oversampling TEchnique (SMOTE) is an over sampling technique based on nearest 
        #neighbors that adds new points between existing points. 
    #Apply imblearn.over_sampling.SMOTE to the dataset. 
    #Build and evaluate the logistic regression model. Is it there any improvement?

from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=50)

X_train_SMOTE, y_train_SMOTE = sm.fit_resample(X_train_scaled, y_train)

classification = LogisticRegression(random_state=0, solver='lbfgs', multi_class = 'ovr' )
classification.fit(X_train_SMOTE, y_train_SMOTE)
y_test_pred = classification.predict(X_test_scaled)

print("The accuracy in the TEST set is: {:.2f}".format(accuracy_score(y_test,y_test_pred)))
print("The precision in the TEST set is: {:.2f}".format(precision_score(y_test,y_test_pred, pos_label="Yes")))
print("The recall in the TEST set is: {:.2f}".format(recall_score(y_test,y_test_pred, pos_label="Yes")))
print("The F1 in the TEST set is: {:.2f}".format(f1_score(y_test,y_test_pred, pos_label="Yes")))
   
```

```python
# We've got small improvements in F and some inmprovements in recall. 
    #Although the accuracy is now a little lower but the precision decreases a lot.
    #In general we can assume that we do not get improvements, in this model, when SMOTE technique is applied.
    
```
